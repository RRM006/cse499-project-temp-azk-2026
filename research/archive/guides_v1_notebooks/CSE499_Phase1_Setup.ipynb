## STEP 5 — ✅ Verify Setup: Print Full Folder Tree

This prints everything that was created so you can confirm it matches the plan.

````
def print_tree(path, prefix='', max_depth=4, current_depth=0):
    if current_depth > max_depth:
        return
    if not os.path.isdir(path):
        return
    entries = sorted(os.listdir(path))
    for i, entry in enumerate(entries):
        full = os.path.join(path, entry)
        connector = '└── ' if i == len(entries) - 1 else '├── '
        icon = '📁' if os.path.isdir(full) else '📄'
        print(f'{prefix}{connector}{icon} {entry}')
        if os.path.isdir(full):
            extension = '    ' if i == len(entries) - 1 else '│   '
            print_tree(full, prefix + extension, max_depth, current_depth + 1)

print(f'📁 CSE499_EHR_Project/')
print_tree(BASE)
````

## 💾 CHECKPOINT GUIDE — How to Never Lose Training Progress

**The Problem:** Colab free tier disconnects after ~4 hours. Without checkpoints, all training is lost.

**The Solution:** Save a checkpoint to Google Drive every 500 steps.
When Colab disconnects → reload notebook → mount Drive → resume from last checkpoint.

---
