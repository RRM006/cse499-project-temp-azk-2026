This notebook creates the **complete project folder structure** inside your Google Drive. Every folder from the DOCX guide is created here — but **only the Phase 1 folders are populated now**. Phase 2 and Phase 3 folders are created empty so the structure is ready when you get there.

## Why Google Drive?
Google Colab's local storage is **temporary** — it gets wiped when the session ends or disconnects. By working inside Google Drive:
- Your **datasets**, **model checkpoints**, **scripts**, and **results** persist across sessions
- If Colab disconnects mid-training, your saved checkpoints are safe
- You can resume from exactly where you stopped

## What gets saved where?
| Item | Location | Why |

| Raw downloaded audio | `raw_audio/dhaka/`, `raw_audio/chittagong/` | Source files, need to keep |

| CSV datasets | `dataset/` | Train/val/test splits |
| Model checkpoints | `models/` | **CRITICAL** — training progress lives here |
| Python scripts | `scripts/` | Reusable code files |
| WER scores / results | `results/` | Evaluation output |
| Colab notebooks | `notebooks/` | Your working notebooks |

## Step 4: Verify the Folder Structure
This prints the full tree so you can visually confirm everything is correct.

````python
# ════════════════════════════════════════════════════════════════
# VERIFY: Print full folder tree
# ════════════════════════════════════════════════════════════════

def print_tree(path, prefix=''):
    entries = sorted(os.listdir(path))
    dirs = [e for e in entries if os.path.isdir(os.path.join(path, e))]
    files = [e for e in entries if os.path.isfile(os.path.join(path, e))]

    for f in files:
        print(f'{prefix}📄 {f}')

    for i, d in enumerate(dirs):
        is_last = (i == len(dirs) - 1) and (len(files) == 0 or i == len(dirs) - 1)
        connector = '└── ' if is_last else '├── '
        print(f'{prefix}{connector}📁 {d}/')
        extension = '    ' if is_last else '│   '
        print_tree(os.path.join(path, d), prefix + extension)

print(f'📂 NSU_499A/')
print_tree(BASE)
````
