##  Create all folders


```python
created_count = 0
skipped_count = 0

for folder in folders:
    if not os.path.exists(folder):
        os.makedirs(folder, exist_ok=True)
        created_count += 1
        print(f'✅ Created: {folder.replace(ROOT, "CSE499_EHR_Project")}')
    else:
        skipped_count += 1
        print(f'⏭️  Already exists: {folder.replace(ROOT, "CSE499_EHR_Project")}')

print(f'\n--- Done: {created_count} created, {skipped_count} already existed ---')
```

## Create the `dataset_log.csv` template
This CSV file tracks every audio file. The columns are: filename, source_url, dialect, speaker_gender, approximate_age, duration_seconds.

```python
import csv

csv_path = f'{ROOT}/01_Dataset/metadata/dataset_log.csv'

if not os.path.exists(csv_path):
    with open(csv_path, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([
            'filename',
            'source_url',
            'dialect',
            'speaker_gender',
            'approximate_age',
            'duration_seconds'
        ])
    print(f'✅ Created: dataset_log.csv')
else:
    print(f'⏭️  dataset_log.csv already exists')
```

## Create the EHR JSON template
This is the exact EHR template.This template and fill in the fields using NER output. We create it now so the structure is available from the start.

```python
import json

ehr_template = {
    "patient_id": "",
    "recording_date": "",
    "dialect": "",
    "audio_file": "",
    "transcript": "",
    "chief_complaint": "",
    "symptoms": [
        {"name": "", "severity": "", "duration": ""}
    ],
    "diseases_mentioned": [],
    "medications": [
        {"name": "", "dosage": "", "frequency": ""}
    ],
    "allergies": [],
    "body_parts_affected": [],
    "duration_of_illness": "",
    "doctor_notes": "",
    "ehr_generated_at": ""
}

template_path = f'{ROOT}/04_Phase3_EHR/templates/ehr_template.json'

if not os.path.exists(template_path):
    with open(template_path, 'w', encoding='utf-8') as f:
        json.dump(ehr_template, f, ensure_ascii=False, indent=2)
    print(f'✅ Created: ehr_template.json')
else:
    print(f'⏭️  ehr_template.json already exists')

print('\n📋 Template preview:')
print(json.dumps(ehr_template, ensure_ascii=False, indent=2))
```

## Verify the complete folder tree


```python
print('CSE499_EHR_Project/')
for dirpath, dirnames, filenames in sorted(os.walk(ROOT)):
    # Calculate depth for indentation
    depth = dirpath.replace(ROOT, '').count(os.sep)
    indent = '│   ' * depth
    basename = os.path.basename(dirpath)
    if dirpath == ROOT:
        continue
    print(f'{indent}├── {basename}/')
    # Show files in this directory
    sub_indent = '│   ' * (depth + 1)
    for filename in sorted(filenames):
        print(f'{sub_indent}├── {filename}')

print('\n✅ Setup complete')
```
